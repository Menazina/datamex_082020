```python
new_cust['Type of customer']=np.where(new_cust['amount_spent']>new_cust['amount_spent'].quantile(.95),'VIP',
                                        np.where(new_cust['amount_spent']>new_cust['amount_spent'].quantile(.75),'Preferred','nan'))
```

```python
def label(y):
    if y>new_cust['amount_spent'].quantile(.95):
        return 'VIP'
    elif y>new_cust['amount_spent'].quantile(.75):
        return 'Preferred'
    else:
        return 'simple'
    
orders['Type customer']=new_cust['amount_spent'].apply(label)
```

```python
orders[orders['Type customer']=='VIP'].head()
```

```python
l=orders[['Country','Type customer']][orders['Type customer']=='VIP']
maximo=l.groupby('Country').count().max()
l3=l.groupby('Country').count()
l3=l3.sort_values('Type customer',ascending=False)
l3
l3.index[0]
```

```python
l=orders[['Country','Type customer']][(orders['Type customer']=='VIP') | (orders['Type customer']=='Preferred')]
maximo=l.groupby('Country').count().max()
l3=l.groupby('Country').count()
l3=l3.sort_values('Type customer',ascending=False)
l3.index[0]
```
