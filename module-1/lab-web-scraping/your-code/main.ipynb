####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url_7 = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
req_7 = req.get(url_7)
soup_7 = bs(req_7.content,'html.parser')
titles = ['Date & Time UTC', 'Latitude degrees','Longitude degrees','Region name']
sismi = soup_7.find_all('td',{'class':'tabev6'})
lat = soup_7.findAll('td',{'class':'tabev1'})
lat1 = soup_7.findAll('td',{'class':'tabev2'})
reg = soup_7.find_all('td',{'class':'tb_region'})
total = []
````

````python
for j in range(20):
    hours = (sismi[j].text.replace('earthquake','').replace('\xa0\xa0\xa0',' '))
    lati = (lat[2*j+1].text.replace('\xa0','')+' '+lat1[3*j+1].text.replace('\xa0\xa0',''))
    long = (lat[2*j].text.replace('\xa0','')+' '+lat1[3*j].text.replace('\xa0\xa0',''))
    zone = (reg[j].text.replace('\xa0',''))
    total.append([hours,lati,long,zone])
````

````python
pd.DataFrame(total, columns=titles)
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url_10 = 'https://www.wikipedia.org/'
````

````python
#your code
req_10 = req.get(url_10)
soup_10 = bs(req_10.content,'html.parser')
````

````python
lang = soup_10.find('div',{'class':'central-featured'}).find_all('div')
print(lang[0].text)
languages= []
for k in lang:
    lan = re.findall(r'\w+',k.text)[0]
    art = ''.join(re.findall(r'\d+',k.text))
    languages.append([lan, art])
languages
````

#### IMDB's Top 250 data (movie name, Initial release, director name and stars) as a pandas dataframe

````python
# This is the url you will scrape in this exercise 
url_13 = 'https://www.imdb.com/chart/top'
````

````python
# your code
req_13 = req.get(url_13)
soup_13 = bs(req_13.content,'html.parser')
````

````python
movie = []
films = soup_13.find('tbody',{'class':'lister-list'}).find_all('td')
for l in range(249):
    name=re.findall(r'\w+ ?\w+ ?\w+ ?\w+ ?\w ?\w',films[5*l+1].text)
    date=re.findall(r'\(\d+',films[5*l+1].text)[0].replace('(','')
    stars=films[5*l+2].text.replace('\n','')
    movie.append([name, date, stars])

movie
````
